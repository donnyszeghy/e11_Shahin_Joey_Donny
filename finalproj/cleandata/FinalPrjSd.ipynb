```python
MCS1 = np.mean(D11['counts'])
MCS2 = np.mean(D12['counts'])
MCS3 = np.mean(D13['counts'])
MCS4 = np.mean(D14['counts'])
MCS5 = np.mean(D15['counts'])
MAS1 = np.mean(D11['altitude'])
MAS2 = np.mean(D12['altitude'])
MAS3 = np.mean(D13['altitude'])
MAS4 = np.mean(D14['altitude'])
MAS5 = np.mean(D15['altitude'])
MOS1 = np.mean(D6['co2'])
MOS2 = np.mean(D7['co2'])
MOS3 = np.mean(D8['co2'])
MOS4 = np.mean(D9['co2'])
MOS5 = np.mean(D10['co2'])
S_counts = [MCS1,MCS2,MCS3,MCS4,MCS5]
S_altitude = [MAS1,MAS2,MAS3,MAS4,MAS5]
S_co2 = [MOS1,MOS2,MOS3,MOS4,MOS5]
```

```python
def Graphing(x_data1, y_data1, x_data2, y_data2, x_data3, y_data3, x_label, y_label, title,):
    plt.figure(figsize=(10, 6))  # Creates a new figure with a specified size
    plt.plot(x_data1, y_data1, marker='o', label='N Series')  # Plot first series
    plt.plot(x_data2, y_data2, marker='x', label='M Series')  # Plot second series
    plt.plot(x_data3, y_data3, marker='^', label='S Series')  # Plot third series
    plt.title(title)  # Set the title of the graph
    plt.xlabel(x_label)  # Set the label for the x-axis
    plt.ylabel(y_label)  # Set the label for the y-axis
    plt.legend()  # Display legend
    plt.grid(True)  # Enable grid
    plt.show()
 



# Now, call the Graphing functio
Graphing(N_altitude, N_counts, M_altitude, M_counts, S_altitude, S_counts, 'Altitude (meters)', 'Counts', 'Counts vs Altitude Comparison')
```
