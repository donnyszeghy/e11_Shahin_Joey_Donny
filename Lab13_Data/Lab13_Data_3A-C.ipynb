```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

D1 = pd.read_csv('Loc1_session1_group1_2024-04-10_16-06-56.csv') 
D2 = pd.read_csv('Loc1_session2_group1_2024-04-10_16-18-39.csv') 
D3 = pd.read_csv('Loc1_session3_group1_2024-04-10_16-28-32.csv')
```

```python
def Graphing(D,str1,str2):
    plt.plot(D)
    plt.title(str1 + ' vs Time ' + str2)
    plt.xlabel('Time')
    plt.ylabel(str1)
    plt.show()
    
    plt.hist(D)
    plt.title('Distribution of ' + str1)
    plt.xlabel(str1)
    plt.ylabel('Frequency')
    plt.show()
    
    Mean = np.mean(D)
    STD = np.std(D)
    CSTD = np.sqrt(Mean)
    RatioS = STD/CSTD
    
    print('The mean is',Mean)
    print('The STD is', STD)
    print('The STD using the mean as the expected counts is',CSTD)
    print('The ratio of the actual STD over counting statistics STD is', RatioS)
```

```python
Graphing(D1['counts'],'Counts','Jacobs Hall')
```

```python
Graphing(D1['Temperature'],'Temperature','Jacobs Hall')
```

```python
Graphing(D1['Humidity'],'Humidity','Jacobs Hall')
```
