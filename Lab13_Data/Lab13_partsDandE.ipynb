```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math 

D1 = pd.read_csv('Loc1_session1_group1_2024-04-10_16-06-56.csv') 
D2 = pd.read_csv('Loc1_session2_group1_2024-04-10_16-18-39.csv') 
D3 = pd.read_csv('Loc1_session3_group1_2024-04-10_16-28-32.csv')
def Graphing(D,str1,str2):
    plt.plot(D)
    plt.title(str1 + ' vs Time ' + str2)
    plt.xlabel('Time')
    plt.ylabel(str1)
    plt.show()
    
    plt.hist(D)
    plt.title('Distribution of ' + str1)
    plt.xlabel(str1)
    plt.ylabel('Frequency')
    plt.show()
    
    Mean = np.mean(D)
    STD = np.std(D)
    CSTD = np.sqrt(Mean)
    RatioS = STD/CSTD
    
    print('The mean is',Mean)
    print('The STD is', STD)
    print('The STD using the mean as the expected counts is',CSTD)
    print('The ratio of the actual STD over counting statistics STD is', RatioS)
Graphing(D1['counts'],'Counts','Jacobs Hall')
Graphing(D1['Temperature'],'Temperature','Jacobs Hall')
Graphing(D1['Humidity'],'Humidity','Jacobs Hall')
```

```python
N1 = math.sqrt(sum(D1['counts']))
N2 = math.sqrt(sum(D2['counts']))
N3 = math.sqrt(sum(D3['counts']))
NS1 = math.sqrt(len(D1['counts']))
NS2 = math.sqrt(len(D2['counts']))
NS3 = math.sqrt(len(D3['counts']))
Unc_Jac = N1/(math.sqrt(2*NS1))
Unc_Cor = N2/(math.sqrt(2*NS2))
Unc_Bet = N3/(math.sqrt(2*NS3))
```

```python
print(Unc_Jac,"This is the uncertainty on your measured standard deviation from the central limit theorem for Jacobs")
print(Unc_Cor,"This is the uncertainty on your measured standard deviation from the central limit theorem for Cory")
print(Unc_Bet,"This is the uncertainty on your measured standard deviation from the central limit theorem for Betchel")
```

We immediately see why the uncertainities are so large, because our observed and expected variances have large differences, but when calculating the UNC on the STD we need to square it to get the UNC on the variance, doing this all of our values are well within the uncertainity range

Some immediate things to point out are the Number of Data points that were recorded at Jacobs were higher and therefore the counts are higher at jacobs which explains it's higher value of uncertatinity.
