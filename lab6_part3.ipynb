```python
OG1 = np.mean(outside1["Gas"])
OG2 = np.mean(outside2["Gas"])
StdG1 = np.std(outside1["Gas"])



OA1 = np.mean(outside1["Altitude"])
OA2 = np.mean(outside2["Altitude"])
StdA1 = np.std(outside1["Altitude"])

OP1 = np.mean(outside1["Pressure"])
OP2 = np.mean(outside2["Pressure"])
StdP10 = np.std(outside1["Pressure"])

OPm25 = np.mean(outside1["pm25"])
O2Pm25 = np.mean(outside2["pm25"])
StdPm25 = np.std(outside1["pm25"])

OPm10 = np.mean(outside1["pm10"])
O2Pm10 = np.mean(outside2["pm10"])
StdPm10 = np.std(outside1["pm10"])

OH1 = np.mean(outside1["Humidity"])
OH2 = np.mean(outside2["Humidity"])
StdH1 = np.std(outside1["Humidity"])

OT1 = np.mean(outside1["Temperature"])
OT2 = np.mean(outside2["Temperature"])
StdT1 = np.std(outside1["Temperature"])

#InsideD = (InM)
N = math.sqrt(270)
UnG = (StdG1/(N))*3      
UnA = (StdA1/(N))*3
UnP = (StdP1/(N))*3
Unp25 = (StdPm25/(N))*3
Unp10 = (StdPm10/(N))*3
UnH = (StdH1/(N))*3
UnT = (StdT1/(N))*3

CompA = abs(OA1-OA2)
CompG = abs(OG1-OG2)
CompP = abs(OP1-OP2)
Comp25 = abs(OPm25-O2Pm25)
Comp10 = abs(OPm10 - O2Pm10)
CompH = abs(OH1 - OH2)
CompT = abs(OT1 - OT2)

if CompG > UnG:
    print('data is statistically different for Gas')
else: 
    print ('data is not statistically different for Gas')
if CompA > UnA:
    print('data is statistically different for Altitude')
else: 
    print ('data is not statistically different for Altitude')
if CompP > UnP:
    print('data is statistically different for Pressure')
else: 
    print ('data is not statistically different for Pressure')
if Comp25 > Unp25: 
    print('data is statistically different for pm25')
else: 
    print ('data is not statistically different for pm25')
if Comp10 > Unp10:
    print('data is statistically different for pm10')
else: 
    print ('data is not statistically different for pm10')
if CompH > UnH:
    print('data is statistically different for Humidity')
else: 
    print ('data is not statistically different for Humidity')
if CompT > UnT:
    print('data is statistically different for Temperature')
else: 
    print ('data is not statistically different for Temperature')
```
