```python
import pandas as pd
import matplotlib.pyplot as plt
import csv
import numpy as np
```

```python
df3 = pd.read_csv('lab4inside.csv')
df2 = pd.read_csv("lab4outside.csv") #our data
df4 = pd.read_csv('insidedata.csv')
df1 = pd.read_csv("outsidedata.csv") #other groups data
new_data = df1 
new_data2 = df2
us = 'Our Data'
them = "Other Group's Data"

def plotcomps(ourname, ourfile, theirname, theirfile, title):
    us = 'Our Data'
    them = "Other Group's Data"
    plt.hist(theirfile[theirname][30:], label = them)
    plt.hist(ourfile[ourname][30:], label = us)
    plt.title(title)
    plt.legend()
    plt.show()
```

```python
plotcomps('pm25', df2, 'PM25', df1, 'PM25 Outside')
plotcomps('pm25', df3, 'PM25', df4, 'PM25 Inside')
```
