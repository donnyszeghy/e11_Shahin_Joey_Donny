```python
import pandas as pd
import matplotlib.pyplot as plt
import csv
import numpy as np
```

```python
df1 = pd.read_csv("d2.csv")
new_data = df1.loc[:,"Time":"Gas"]
new_data
```

```python
def compgraph(x, y):
    plt.plot(new_data[x], new_data[y])
    plt.title(x +'vs'+ y)
    plt.show()
```

```python
compgraph('Temperature', 'Humidity')
```

The hidden variable here when making plots like Temp vs Humid is the time, we can't have a correlation of data when they are taken at different times. The time is what allows us to correlate the data between the two

The plot shows that when we got outside Temperature and Humidity are linearly correlated. The pearsons coeffiecent shows us how correlated 2 sets of data are and we can indeed use mathmatical functions in code to implement this it would be 1 - resisdual sum of residuals/total sum of squares of the residual

```python
compgraph('Temperature', 'Pressure')
```

There seems to be a negative linear correlation in this graph - we can use PV=NRT to see this 

```python
compgraph('Temperature', 'pm10')
```
